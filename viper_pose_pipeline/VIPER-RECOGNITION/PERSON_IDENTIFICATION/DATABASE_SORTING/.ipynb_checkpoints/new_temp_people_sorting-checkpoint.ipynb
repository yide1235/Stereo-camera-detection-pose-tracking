```python
def binary_search(arr,value):
    left=0
    right=len(arr)-1
    
    if right<0:
        return 0
    
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==value:
            return mid
        if arr[mid]<value:
            left=mid+1
        else:
            right=mid-1
            
    return mid

def filter_global_data(arr,lookup,percentage):
    index_matched=[]
    list_inds=[]
    started=False
    for i in range(len(lookup)):
        if lookup[i]==0:
            list_inds.append([])
            continue
        lower=lookup[i]*(1-percentage)
        upper=lookup[i]*(1+percentage)
        lower_ind=binary_search(arr[:,i*2+1],lower)
        upper_ind=binary_search(arr[:,i*2+1],upper)
        
        if arr[j,i*2+1]==0:
            j=0
            while True:
                if arr[j,i*2+1]==0:
                    pass
                else:
                    last_0=j
                    break
                j=j+1
        else:
            last_0=-1
            
        
        if len(index_matched)==0 and started==False:
            index_matched=arr[lower_ind:upper_ind+1,i*2]
            if last_0 != -1:
                index_matched=index_matched+list(range(last_0))
            started=True
            list_inds.append([lower_ind,upper_ind+1,last_0])
        else:
            index_matched=list(set(index_matched).intersection(arr[lower_ind:upper_ind+1,i*2]))
            if last_0 != -1:
                index_matched=index_matched+list(range(last_0))
            list_inds.append([lower_ind,upper_ind+1,last_0])
    return index_matched, list_inds
```

```python
class Frame_data:
    def __init__(self, frame_id, data, bounding_box, face_box, face_score, daytime):
        self.id = frame_id
        self.data = data
        self.bounding_box=bounding_box
        self.face_box=face_box
        self.face_score=face_score
        self.daytime=daytime
        
```

```python
class Global_person:
    def __init__(self,person):
        self.count=1
        self.global_id=person.global_id
        self.frame_ids=person.frame_ids
        self.face_frame=person.face_frame
        self.bounding_box=person.bounding_box
        self.face_box=person.face_box
        self.face_score=person.face_score
```

```python
class Person:
    def __init__(self,temp_id,vec,frame_id,maturity_threshold,bounding_box,face_box,face_score,daytime):
        self.temp_id=temp_id
        self.global_id=None
        self.frame_ids=[frame_id]
        self.face_frame=frame_id
        self.face_frame=face_box
        self.bounding_box=[bounding_box]
        self.face_score=face_score
        self.mature=False
        self.maturity_cutoff=maturity_threshold
        self.data=[[0] for _ in range(len(vec))]
        self.mean=[0]*len(vec)
        self.lastdt=daytime
        
        for i in range(len(vec)):
            self.data[i]=[vec[i]]
            self.mean[i]=vec[i]
        
    def add(self,vec,frame_id,bounding_box,face_box,vec_threshold,face_score,global_search_percentage,daytime,iqr_cutoff,max_bad_vecs):
        if self.mature==False:
            for i in range(len(vec)):
                if vec[i] != 0:
                    if self.data[i][0]!=0:
                        self.data[i].append(vec[i])
                    else:
                        self.data[i]=[vec[i]]

            self.bounding_box.append(bounding_box)
            if face_score>self.face_score:
                self.face_frame=frame_id
                self.face_box=face_box
                self.face_score=face_score
            self.face_box.append(bounding_box)
            self.frame_ids.append(frame_id)
            self.update_mean()
            self.lastdt=daytime
            self.check_maturity(vec_threshold,face_threshold,global_search_percentage,iqr_cutoff,max_bad_vecs)
        else:
            if face_score>self.face_score:
                self.face_frame=frame_id
                self.face_box=face_box
                self.face_score=face_score
            self.bounding_box.append(bounding_box)
            self.face_box.append(bounding_box)
            self.frame_ids.append(frame_id)
            self.lastdt=daytime
                
                    
    def update_mean(self):
        for i in range(len(self.data)):
            self.mean[i]=np.mean(self.data[i])
            
    def check_range(self,iqr_cutoff,max_bad_vecs):
        bad_inds=[]
        for i in range(len(self.data)):
            #NEED TO MODIFY THIS CONDITION TO DESIRED CHECK
            q1 = np.percentile(self.data[i], 25)
            q3 = np.percentile(self.data[i], 75)
            iqr = q3 - q1
            if iqr>iqr_cutoff*self.mean[i]:
                bad_inds.append(i)
        if len(bad_inds)>max_bad_vecs:
            return False
        else:
            self.mature_data=[]
            for i in range(len(self.data)):
                if i in bad_inds:
                    self.mature_data.append(0)
                else:
                    self.mature_data.append(self.data[i])
            return true
                    
    def check_maturity(self,vec_threshold,face_threshold,global_search_percentage,iqr_cutoff,max_bad_vecs):
        if self.mature!=True:
            mature=True
            for i in range(len(self.data)):
                if len(self.data[i])<self.maturity_cutoff:
                    mature=False
            if mature:
                if not self.check_range(iqr_cutoff,max_bad_vecs):
                    mature=False
                else:
                    self.mature=True
                    self.get_global_id(vec_threshold,face_threshold,global_search_percentage)
            return mature
        else:
            return True
            
    def check_similarity(self, candidate_vec, threshold, max_non_matches):
        count=0
        for i in range(len(self.mean)):
            if candidate_vec[i]==0 or self.mean[i]==0:
                count=count+1
                continue
            if abs(self.mean[i]-candidate_vec[i])>threshold:
                count=count+1
            if count>max_non_matches:
                return False
        return True
    
    def get_global_id(self,vec_threshold,face_threshold,global_search_percentage):
        good=[]
        score=[]
        global global_people_data
        global global_people_list
        
        if len(global_people_list)==0:
            self.global_id=get_next_global_id()
            self.add_to_global_dataset()
            return
            
        
        found_ids, list_inds=filter_global_data(global_people_data,self.mature_data,global_search_percentage)
        
        found=[]
        for i in found_ids:
            found.append(global_people_list[i])
        
        for person in found:
            face_similarity=run_face_check(person.face_frame[0],self.face_frame[0])
            if face_similarity < face_threshold:
                good.append(person)
                score.append(face_similarity)
                
        if len(good)==0:
            self.global_id=get_next_global_id()
            self.add_to_global_dataset()
        elif len(good)==1:
            self.global_id=good[0].global_id
            self.update_global_dataset(list_inds)
        else:
            self.global_id=good[score.index(max(score))].global_id
            self.update_global_dataset(list_inds)
            
    def update_global_dataset(self, list_inds):
        global global_people_data
        global global_people_list
        
        global_person=global_people_list[self.global_id]
        
        global_person.count=global_person.count+1
        
        for i in range(len(self.mature_data)):
            
            if self.mature_data[i]==0:
                continue
                
            try:
                old_ind=global_people_data[list_inds[i][0],list_inds[i][1],i*2].index(self.global_index)+list_inds[i][0]
            except:
                old_ind=global_people_data[0,list_inds[i][2],i*2].index(self.global_index)
                
            old_val=global_people_data[old_ind,i*2]
            if old_val==0:
                new_val=self.mature_data[i]
            else:
                new_val=((global_person.count-1)*old_val+self.mature_data[i])/global_person.count
            new_ind=binary_search(global_people_data[:,i*2+1],self.mature_data[i])
            global_people_data[:,i*2].pop(old_ind)
            global_people_data[:,i*2+1].pop(old_ind)
            global_people_data[:,i*2]=np.insert(global_people_data[:,i*2],ind,self.global_id)
            global_people_data[:,i*2+1]=np.insert(global_people_data[:,i*2+1],ind,new_val)
        
            
    def add_to_global_dataset(self):
        global global_people_data
        global global_people_list            
        
        if len(global_people_list)>self.global_id:
            global_people_list[self.global_id]=Global_person(self)
        elif len(global_people_list)==self.global_id:
            global_people_list.append(Global_person(self))
        
        for i in range(len(self.mature_data)):
            ind=binary_search(global_people_data[:,i*2+1],self.mature_data[i])
            global_people_data[:,i*2]=np.insert(global_people_data[:,i*2],ind,self.global_id)
            global_people_data[:,i*2+1]=np.insert( global_people_data[:,i*2+1],ind,self.mature_data[i])
    
                
```

```python
def get_next_temp_id():
    global next_id
    id_val=next_id
    next_id=next_id+1
    return id_val

def get_next_global_id():
    global next_global_id
    id_val=next_global_id
    next_global_id=next_global_id+1
    return id_val

def run_face_check(face1_frame,face2_frame):
        return next_id % 2 == 0
    
def secondary_check(vec,person,threshold):
    score=get_euclidean_score(vec,person)
    if score<threshold:
        return True
    else:
        return False

def get_euclidean_score(vec,person):
    vec1=person.mean
    count1=0
    dist1=0
    for i in range(len(vec)):
        if vec[i] !=0:
            if vec1[i] !=0:
                count1=count1+1
                dist1=dist1+(vec[i]-vec1[i])**2
                
    dist1=dist1/count1
    dist1=dist1/(1.05**count1)
    
    return dist1

def euclidean_check(vec, found):
    scores=[]
    for i in range(len(found)):
        scores.append(get_euclidean_score(vec,found[i]))
        
    best_ind=scores.index(max(scores))
    return found[best_ind]

def clear_temp_list(temp_list,hold_time):
    for person in list(temp_list):
        if (person.lastdt - datetime.now()) > timedelta(seconds=hold_time):
            temp_list.remove(person)
    return temp_list
    
def sort_frame_data(frame,people_list,soft_similarity_threshold,hard_similarity_threshold,maturity_threshold,vec_threshold,face_threshold,secondary_euclidean_threshold,global_search_percentage,temp_list_hold_time,max_non_matches,iqr_cutoff,max_bad_vecs,zero_tol):
    people_list=clear_temp_list(people_list,temp_list_hold_time)
    for i in range(len(frame.data)):
        found=[]
        person_vec=frame.data[i]
        count=0
        for j in range(len(person_vec)):
            if person_vec[j]>0:
                count=count+1
        if count<zero_tol:
            continue
        for person in people_list:
            hard_similarity=person.check_similarity(person_vec,hard_similarity_threshold,max_non_matches)
            if hard_similarity==True:
                found.append(person)
            else:
                soft_similarity=person.check_similarity(person_vec,soft_similarity_threshold,max_non_matches)
                print("soft,",soft_similarity)
                if soft_similarity==True:
                    if secondary_check(person_vec,person,secondary_euclidean_threshold)==True:
                        print("secondary,True")
                        found.append(person)
        if len(found)==0:
            print("Found 0")
            next_id=get_next_temp_id()
            people_list.append(Person(next_id,person_vec,frame.id,maturity_threshold,frame.bounding_box[i],frame.face_box[i],frame.face_score[i],frame.daytime)) 
        elif len(found)==1:
            print("Found 1")
            found[0].add(person_vec,frame.id,frame.bounding_box[i],frame.face_box[i],vec_threshold,frame.face_score[i],global_search_percentage,frame.daytime,iqr_cutoff,max_bad_vecs)
        else:
            print("Found 2+")
            best=euclidean_check(person_vec,found)
            best.add(person_vec,frame.id,frame.bounding_box[i],frame.face_box[i],vec_threshold,frame.face_score[i],global_search_percentage,frame.daytime,iqr_cutoff,max_bad_vecs)
    return people_list
            
            
```
